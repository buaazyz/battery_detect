```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('small.png') 
print(img.shape)
plt.imshow(img)
#pltcv2 RGB
```

```python
def ReSizeInput(img,size = (1040,2000,3)):
    if img.shape<=size:
        img_new = cv2.copyMakeBorder(img,0,size[0]-img.shape[0],0,size[1]-img.shape[1],cv2.BORDER_CONSTANT,value=[255,255,255])
    else:
        img_new = img[:size[0],:size[1],:]
    #print(img_new.shape)
    return img_new
```

```python
img = cv2.imread('small.png') 

plt.imshow(ReSizeInput(img))
```

```python
img = cv2.imread('large1.png') 

plt.imshow(ReSizeInput(img))
```
